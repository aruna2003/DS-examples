# GROUPING AND AGGREGATING

```python
sold_price_by_age = ipldf.groupby('AGE')['SOLD PRICE'].mean().reset_index()
sold_price_by_age
```

# ADDING AND REMOVING COLUMNS

```python
ipldf['PREMIUM']=ipldf['SOLD PRICE']-ipldf['BASE PRICE']
ipldf[['PLAYER NAME','BASE PRICE','SOLD PRICE','PREMIUM']][0:5]
```

```python
#ipldf.shape
ipldf = ipldf.drop('ECON',axis = 1)
ipldf.shape
```

# MERGING DATAFRAME

```python
sold_price_by_age_role = ipldf.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
sold_price_by_age_role
```

```python
soldprice_comparison=sold_price_by_age_role.merge(sold_price_by_age,on = 'AGE')
```

```python
soldprice_comparison
```

```python
soldprice_comparison.rename(columns = 
{'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'},
                            inplace = True)
```

```python
soldprice_comparison
```

# FILTERING

```python
soldprice_comparison['CHANGE']=soldprice_comparison.apply(lambda x:(x.SOLD_PRICE_AGE_ROLE-x.SOLD_PRICE_AGE)/x.SOLD_PRICE_AGE,axis = 1)
soldprice_comparison
```

```python
soldprice_comparison[soldprice_comparison.CHANGE > 0]
```

# HANDLING NULL VALUES

```python
soldprice_comparison[soldprice_comparison.CHANGE < 0]= None
soldprice_comparison
```

```python
soldprice_comparison[soldprice_comparison.CHANGE.isnull()]
```

```python
soldprice_comparison = soldprice_comparison.dropna(subset = ['CHANGE'])
```

```python
soldprice_comparison[soldprice_comparison.CHANGE.isnull()]
```
