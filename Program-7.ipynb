Develop a program to implement Naive Bayes classifier model and analyze the model using confusion matrix

```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
```

```python
iris_df = pd.read_csv('Iris_data_sample.csv',header =None)
iris_df.info()
iris_df
```

```python
iris_df = iris_df.replace(to_replace =['??','###'],value =None) 
iris_df.dropna(axis = 0, how='any',inplace = True)
iris_df.info()
iris_df.iloc[:,5].value_counts()
```

```python
iris_df
```

```python
X_features = iris_df.iloc[:,1:5]
X_features
```

```python
Y_features = iris_df.iloc[:,5]
Y_features
```

```python
model = GaussianNB()
model.fit(X_features, Y_features)
```

```python
expected = Y_features
predicted = model.predict(X_features)
predicted
```

```python
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(expected, predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```

```python
#Precision = TP/(TP + FP),
#Recall = TP/(TP+FN),
#F1 Score = 2*(Recall * Precision) / (Recall + Precision),
#the number of actual occurrences of the class in the specified dataset
```

```python
print(metrics.classification_report(expected, predicted))
```
