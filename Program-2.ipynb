Using pandas and Matplotlib demonstrate the following operations for the sample dataset given, i) Bar chart and Histogram ii) Comparing Distribution iii) Box plot and mention quartiles iv) Correlation using pairplot and heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ipldf = pd.read_csv('IPLDATA.csv')
ipldf.head()
```

BAR CHART

```python
sns.barplot(x='AGE',y='SOLD PRICE',data = ipldf)   #represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
```

```python
sns.barplot(x='AGE',y='SOLD PRICE',data = ipldf,hue='PLAYING ROLE')
```

HISTOGRAM

```python
plt.hist(ipldf['SOLD PRICE']) #a graphical representation of data points organized into user-specified ranges
```

```python
plt.hist(ipldf['SOLD PRICE'],bins = 50)
```

DIstribution or Density Chart

```python
sns.distplot(ipldf['SOLD PRICE']) # data distribution of a variable against the density distribution
```

Box Plot

```python
box = plt.boxplot(ipldf['SOLD PRICE'])
```

```python
ipldf[ipldf['SOLD PRICE']>1350000.0][['PLAYER NAME','PLAYING ROLE','SOLD PRICE']]
```

Comparing Distributions

```python
sns.distplot(ipldf[ipldf['CAPTAINCY EXP']==1]['SOLD PRICE'],color='y',label = 'Captaincy Experience')
sns.distplot(ipldf[ipldf['CAPTAINCY EXP']==0]['SOLD PRICE'],color='r',label = 'No Captaincy Experience')
plt.legend()
```

```python
sns.distplot(ipldf[ipldf['CAPTAINCY EXP']==1]['SOLD PRICE'],color='y',label = 'Captaincy Experience')
plt.legend()
```

```python
sns.boxplot(x='PLAYING ROLE',y='SOLD PRICE',data = ipldf)
```

Scatter Plot

```python
ipldf_batsman = ipldf[ipldf['PLAYING ROLE']=='Batsman']
plt.scatter(x=ipldf_batsman.SIXERS,y=ipldf_batsman['SOLD PRICE'])
```

```python
sns.regplot(x='SIXERS',y='SOLD PRICE',data =ipldf_batsman)
```

Pair Plot

```python
infl_features = ['SR-B','AVE','SIXERS','SOLD PRICE']
sns.pairplot(ipldf[infl_features],size = 2)
```

Correlation and HeatMap

```python
ipldf[infl_features].corr()
```

```python
sns.heatmap(ipldf[infl_features].corr(),annot = True)
```
