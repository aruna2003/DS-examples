```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['Department'] = Le.fit_transform(data['Department'])
data['Department'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['Attrition'] = Le.fit_transform(data['Attrition'])
data['Attrition'].unique()
```

```python
## Creating independent and dependent variable
x=data.loc[:,['DailyRate','Education','JobInvolvement','HourlyRate',
            'Department']]    #independent variable
y=data.loc[:,["Attrition"]]   #dependent variable 
print(data['DailyRate'].unique())
print(data['Education'].unique())
print(data['JobInvolvement'].unique())
print(data['HourlyRate'].unique())
print(data['Attrition'].unique())
```

```python
## preparing training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,
                                   test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier     #importing decision tree from sklearn.tree
dt=DecisionTreeClassifier()                  #object creation for decision tree  
dt.fit(X_train,y_train)                     #training the model
y_hat=dt.predict(X_test)                   #prediction
y_hat                                      #predicted values 
```

```python
y_train_predict=dt.predict(y_test)#predicting training data to check training performance 
y_train_predict
```

```python
## Evalauting the model
from sklearn.metrics import accuracy_score,classification_report,f1_score   #importing mertics to check model performance
##Training score
y_train_predict=dt.predict(X_train)   #passing X_train to predict Y_train
acc_train=accuracy_score(y_train,y_train_predict)   #checking accuracy
acc_train
```

```python
print(classification_report(y_train,y_train_predict))# it will give precision,recall,f1 scores and accuracy 
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
print(metrics.confusion_matrix(y_train,y_train_predict))
```
